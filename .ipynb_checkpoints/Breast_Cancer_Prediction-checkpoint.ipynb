## Data Preprocessing 

```python
df.isnull().sum()
```

```python
## In this dataset there is no null values
# the id column is not useful for predicting output so we are dropping that column
df.drop('id',axis=1,inplace=True)
df.head()
```

```python
## Encoding Categorical variable to numericals
## here diagnosis is Categorical
df['diagnosis'].unique()
```

```python
df['diagnosis'] = df['diagnosis'].map({'B':0,'M':1})
df['diagnosis'].unique()
```

## Feature scaling

```python
df.corr()
```

```python
plt.figure(figsize=(25,30))
df_corr = df.corr()
sns.heatmap(df_corr,annot=True,cmap=plt.cm.CMRmap_r)
```

```python

def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: 
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
correlation_columns = correlation(df,0.9)
correlation_columns
```

```python
## Now we are going to drop Correlated Colums
df.drop(correlation_columns,axis=1)
```

## Splitting to Train and Test data

```python
X = df.drop('diagnosis',axis=1)
Y = df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(X,Y, test_size=0.3,random_state=2)
```

```python
print("X_train: ",X_train.shape)
print("X_test: ",X_test.shape)
```

## Standardizing

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
sc.fit_transform(X_train)
```

## Building Machine Learning Model

#  Logistic Regression

### Model Training

```python
from sklearn.linear_model import LogisticRegression
log_regression = LogisticRegression(random_state=5)
log_regression.fit(X_train,y_train)
```

### Model Prediction

```python

y_pred = log_regression.predict(X_test)
y_pred
```

### Model Evaluation

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test,y_pred)
acc
```
